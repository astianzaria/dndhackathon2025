This section contains a trend or insight from the Pie Chart sheet; to see the trend and ask a question:
- What is the busiest area?
- What area has the most damaged carousel? (What is the reason for the damaged carousel?)
- What area has less shelter â€“ does it need some development, if any?

Sheffield got the best overall, while Barnsley stayed the lowest for all parts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from spyder_kernels.utils.iofuncs import load_image

# Read CSV file
try:
    df = pd.read_csv("pie_chart.csv")  
except FileNotFoundError:
    print("Error: File not found. Check the file path.")
    exit()

df.set_index("Council Area", inplace=True)

# Busiest Council Areas
busiest = df.nlargest(4, "Average Boardings Per Hour")[
    ["Average Boardings Per Hour", "Average Buses Per Hour"]
]

# Least Shelter Coverage
least_shelters = df.nsmallest(4, "Stops with Shelters")[
    ["Stops with Shelters"]
]

# damaged Shleter
damaged_shelters = df.nsmallest(4, "Damaged Carousels")[
    ["Damaged Carousels"]
]

# 4. Visualization
plt.figure(figsize=(15, 10))

# Busiest Areas
plt.subplot(2, 2, 1)
sns.barplot(
    x=busiest.index,
    y="Average Boardings Per Hour",
    hue=busiest.index,  # Added hue to fix warning
    data=busiest,
    palette="flare",
    legend=False
)
plt.title("Busiest Areas (Boardings/Hour)")
plt.xticks(rotation=45)

# Least Shelters
plt.subplot(2, 2, 3)
sns.barplot(
    x=least_shelters.index,
    y="Stops with Shelters",
    data=least_shelters,
    palette="flare",
    hue=least_shelters.index,
    legend=False
)
plt.title("Least Shelter Coverage")
plt.xticks(rotation=45)

# Least Shelters
plt.subplot(3, 3, 6)
sns.barplot(
    x=damaged_shelters.index,
    y="Damaged Carousels",
    data=damaged_shelters,
    palette="flare",
    hue=damaged_shelters.index,
    legend=False
)
plt.title("Damaged Shelter")
plt.xticks(rotation=45)

plt.show()
```
